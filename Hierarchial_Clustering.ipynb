```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
!pip install scipy
```

```python
data=pd.read_csv('Wholesale customers data.csv')
data.head()
```

```python
from sklearn.preprocessing import normalize
data_s=normalize(data)
data_s=pd.DataFrame(data_s,columns=data.columns)
data_s.head()
```

```python
import  scipy.cluster.hierarchy as hsc
plt.figure(figsize=(10,7))
plt.title("DendoGrams")
dend=hsc.dendrogram(hsc.linkage(data_s,method='ward'))
```

```python
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
dend=hsc.dendrogram(hsc.linkage(data_s,method='ward'))
plt.axhline(y=6,color='r',linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit_predict(data_s)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_s['Milk'],data_s['Grocery'],c=cluster.labels_)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_s['Frozen'],data_s['Grocery'],c=cluster.labels_)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_s['Milk'],data_s['Detergents_Paper'],c=cluster.labels_)
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

df1['mpg']= label_encoder.fit_transform(df1['mpg'])
df1['displacement']= label_encoder.fit_transform(df1['displacement'])
df1['year']= label_encoder.fit_transform(df1['year'])
df1['weight']= label_encoder.fit_transform(df1['weight'])
df1['cylinders']= label_encoder.fit_transform(df1['cylinders'])
df1['origin']= label_encoder.fit_transform(df1['origin'])
df1['horsepower']= label_encoder.fit_transform(df1['horsepower'])


```

```python
from sklearn.preprocessing import normalize
data_auto=normalize(df1)
data_auto=pd.DataFrame(data_auto,columns=df1.columns)
data_auto.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit_predict(data_auto)
```
